```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import pandas as pd
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

options = webdriver.ChromeOptions()

options.add_experimental_option("detach", True)
options.add_argument("--disable-extensions")
options.add_argument("--disable-notifications")
options.add_argument("--disable-Advertisement")
options.add_argument("--disable-popup-blocking")
options.add_argument("start-maximized")
```

```python
s=Service()
driver= webdriver.Chrome(service=s,options=options)

driver.get('https://www.youtube.com/@tseries/videos')
time.sleep(5)
```

```python
item = []
SCROLL_PAUSE_TIME = 3
last_height =   driver.execute_script("return document.documentElement.scrollHeight")

item_count = 1000

while item_count > len(item):
    driver.execute_script("window.scrollTo(0,100000);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.documentElement.scrollHeight")

    if new_height == last_height:    
        break
    last_height = new_height
data = []
try:
    for e in WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div#details'))):
        vurl = e.find_element(By.CSS_SELECTOR,'a#video-title-link').get_attribute('href')
        date_time = e.find_element(By.XPATH,'.//*[@id="metadata"]//span[@class="inline-metadata-item style-scope ytd-video-meta-block"][2]').text
        data.append({
            'video_url':vurl,
            'date_time':date_time,
            })
except:
    pass
    
item = data
print(item)
print(len(item))

    
```

```python
df = pd.DataFrame(item)
print(df)
```

```python
df['Video_ID'] = df['video_url'].str.extract(r'v=([A-Za-z0-9_-]+)')
```

```python
df
```

```python
def count_letters(string):
    letter_count = {}
    for letter in string:
        if letter.isalpha():  
            letter = letter.lower()  
            if letter in letter_count:
                letter_count[letter] += 1
            else:
                letter_count[letter] = 1
    return letter_count
```

```python
letter_counts = df['Video_ID'].apply(count_letters)
```

```python
total_letter_counts = {}
for counts in letter_counts:
    for letter, count in counts.items():
        if letter in total_letter_counts:
            total_letter_counts[letter] += count
        else:
            total_letter_counts[letter] = count
```

```python
max_letter = max(total_letter_counts, key=total_letter_counts.get)
print(f"{max_letter}{total_letter_counts[max_letter]}.")
```
